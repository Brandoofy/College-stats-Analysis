```python
#open and read
x=open('SP copy.csv')
rdata=x.read()

#packages used
import statistics
import pandas as pn 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data cleaning 
#split at lines, delete quotes and turn into list 
rdata.splitlines()
edata = rdata.replace('"',"")
wdata = edata.replace('\n',",")
prel = list(wdata.split(',')) 
postl = [prel[i:i+8] for i in range(0, len(prel), 8)]  
key=postl.pop(0)
#counts and numbers each list
listcount = 0
while listcount in range(0,len(postl)):
    listcount = listcount + 1
```

```python
#asks and test to see all different values in rows
while True:
    f = input("Do you want to check what values are in a column? \nEnter:'Yes' or 'No' ")
    a = f.casefold()
    if a=="yes":
        t = int(input("Type in a number to see different values in each row \n'0' for gender \n'1' for race\n'2' for parent level of education\n'3' for lunch payments\n'4' for test preparation\n'5' for math score \n'6' for reading score \n'7' for writing score \n"))
        if t == 0:
            rowt = "gender"
        elif t == 1:
            rowt = "race/ethnicity"
        elif t == 2:
            rowt = "parental level of education"
        elif t == 3:
            rowt = "lunch"
        elif t == 4:
            rowt = "test preparation course"
        elif t == 5:
            rowt = "math score"
        elif t == 6:
            rowt = "reading score"
        elif t == 7:
            rowt = "writing score"
        else:
            print("Number is not in range")
            continue 
        diffv = []
        for i in (postl):
            if i[t] not in diffv:
                diffv.append(i[t])
        print('There are',len(diffv),"unique values in",rowt,"they are",diffv)
        continue
    elif a=="no":
        break
    else:
        print("Enter either yes/no")
```

```python
#gender data
fcount = 0 
mcount = 0
for i in (postl):
    if i[0] == 'female':
        fcount = fcount + 1 
    elif i[0] == 'male':
        mcount = mcount + 1 
    else: 
        continue
samplesize= fcount+mcount
allgender = []
for i in (postl):
    allgender.append(i[0])
#gender bar plot
barx = np.array(["Female","Male"])
bary = np.array([fcount,mcount])
tocolors = ["hotpink",'blue']
plt.xlabel("Gender Sample")
plt.bar(barx, bary, color = tocolors)
plt.show()
print('\n','Out of',samplesize,'members','\n','There are',fcount,'Females','and',mcount,'Males')
```

```python
#race group data 
a = 0
b = 0
c = 0
d = 0
e = 0 
for i in (postl):
    if i[1] == 'group A':
        a = a + 1
    elif i[1] == 'group B':
        b = b + 1
    elif i[1] == 'group C':
        c = c + 1
    elif i[1] == 'group D':
        d = d + 1
    elif i[1] == 'group E':
        e = e + 1
    else:
        continue
allrace = []
for i in (postl):
    allrace.append(i[1])
#race group pi chart
racpi = np.array([a, b, c, d, e])
mylabel = ["group A", "group B", "group C", "group D","group E"]
myexplode = [0, 0, 0.2,0,0]
mycolors = ["CadetBlue", "DeepPink", "LawnGreen","Peru",'Maroon']
plt.xlabel("Race Samples")
plt.pie(racpi, labels = mylabel, explode = myexplode, shadow = True,autopct='%1.2f%%',colors = mycolors)
plt.show() 

print('\n',str(round(a/len(postl)*100))+'%','identify with group A','\n',str(round(b/len(postl)*100))+'%','identify with group B','\n',str(round(c/len(postl)*100))+'%','identify with group C','\n',str(round(d/len(postl)*100))+'%','identify with group D','\n',str(round(e/len(postl)*100))+'%','identify with group E')
```

```python
#parent level of education data 
shs = 0
hs = 0
ad = 0
scol = 0
col = 0
mas = 0
for i in (postl):
    if i[2] == 'some high school':
        shs = shs + 1
    elif i[2] == 'high school':
        hs = hs + 1
    elif i[2] == "associate's degree":
        ad = ad + 1
    elif i[2] == 'some college':
        scol = scol + 1
    elif i[2] == "bachelor's degree":
        col = col + 1
    elif i[2] == "master's degree":
        mas = mas + 1
    else:
        continue
allparent = []
for i in (postl):
    allparent.append(i[2])
#parent level of eduation pi chart 
edupi = np.array([shs, hs, ad, scol, col, mas])
mylabels = ["Some High School", "High School", "Associate's Degree", "Some College","Bachelor's Degree","Master's Degree"]
myexplode = [0, 0, 0, 0.2,0,0]
plt.pie(edupi, labels = mylabels, explode = myexplode,shadow = True,autopct='%1.2f%%')
plt.xlabel(" \n Parental Education Data")
plt.show() 
print('\n',str(round(shs/len(postl)*100))+'%','have parents that completed some high school','\n',str(round(hs/len(postl)*100))+'%','have parents that completed high school','\n',str(round(ad/len(postl)*100))+'%',"have parents that completed an associate's degree",'\n',str(round(scol/len(postl)*100))+'%','have parents that completed some college','\n',str(round(col/len(postl)*100))+'%',"have parents that completed an bachelor's degree",'\n',str(round(mas/len(postl)*100))+'%',"have parents that completed an masters's degree")
```

```python
#school lunch data
standard = 0
free = 0
for i in (postl):
    if i[3] == 'standard':
        standard = standard + 1
    elif i[3] == 'free/reduced':
        free = free + 1
    else:
        continue
alllunch = []
for i in (postl):
    alllunch.append(i[3])
#school lunch bar graph        
bars = np.array(['Standard',"Free/Reduced'"])
bart = np.array([standard,free])
luncolor = ["Yellow",'Green']
plt.xlabel("School Lunch Data")
plt.bar(bars, bart, color = luncolor)
plt.show()
print('\n',str(round(free/len(postl)*100))+'%','recive free/reduced lunches','\n',str(round(standard/len(postl)*100))+'%','recive standard priced lunches')
```

```python
#test preparation data
none = 0
completed = 0
for i in (postl):
    if i[4] == 'none':
        none = none + 1
    elif i[4] == 'completed':
        completed = completed + 1
    else:
        continue
allprep = []
for i in (postl):
    allprep.append(i[4])
#test preparation pi chart        
testbars = np.array(['No Test Prep',"Test Prep Completed"])
testbart = np.array([none,completed])
tescolor = ["brown",'gray']
plt.xlabel("Test Preparation Course")
plt.bar(testbars,testbart, color = tescolor)
plt.show()       
print('\n',str(round(completed/len(postl)*100))+'%','have completed test prep','\n',str(round(none/len(postl)*100))+'%','have not completed test prep')
```

```python
#exam scores data

#math score 
matl = []
for i in (postl):
            if i[5] not in matl:
                matl.append(i[5])
allmat = []
for i in (postl):
    matint = int(i[5])
    allmat.append(matint)
mathavg = statistics.mean(allmat)
mathmode = statistics.mode(allmat)
mathmed = statistics.median(allmat)
mathvar = statistics.pvariance(allmat)
mathsd = statistics.stdev(allmat)
#reading score
redl = []
for i in (postl):
            if i[6] not in redl:
                redl.append(i[6])

allred = []
for i in (postl):
    redint = int(i[6])
    allred.append(redint)
readavg = statistics.mean(allred)
readmode = statistics.mode(allred)
readmed = statistics.median(allred)
readvar = statistics.pvariance(allred)
readsd = statistics.stdev(allred)
#writing score
wril = []
for i in (postl):
            if i[7] not in wril:
                wril.append(i[7])
allwri = []
for i in (postl):
    wriint = int(i[7])
    allwri.append(wriint)
wriavg = statistics.mean(allwri)
wrimode = statistics.mode(allwri)
wrimed = statistics.median(allwri)
wrivar = statistics.pvariance(allwri)
wrisd = statistics.stdev(allwri)

#Exam Data box plot 
plt.show()
data_1 = np.random.normal(allmat)
data_2 = np.random.normal(allred)
data_3 = np.random.normal(allwri)
all_data = [data_1, data_2, data_3]
labelsub = ['Math', 'Reading', 'Writing']
box_plt_data=[data_1,data_2,data_3]
plt.title("Exam Scores by Subject")
plt.boxplot(box_plt_data,patch_artist=True,labels=['Math','Reading','Writing'])
plt.show()
print('\n','Math: mean is',str(round(mathavg,2))+'%, median is',str(round(mathmed))+'%, mode is',str(round(mathmode))+'%, variance is',round((mathvar)),'standard deviation is',(round(mathsd,2)),'\n',"Reading: mean is",str(round(readavg,2))+'% median is',str(round(readmed))+'%, mode is',str(round(readmode))+'%, variance is',round((readvar)),'standard deviation is',(round(readsd,2)),'\n',"Writing: mean is",str(round(wriavg,2))+'% median is',str(round(wrimed))+'%, mode is',str(round(wrimode))+'%, variance is',(round(wrivar)),'standard deviation is',(round(wrisd,2)))
```
